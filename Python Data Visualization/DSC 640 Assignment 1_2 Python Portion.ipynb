###### This is the Python portion of assignment 1.2. The below code will bring in a dataset and then provide the following visuals. 

1. 1 bar chart
2. 1 stacked bar chart
3. 1 pie chart
4. 1 donut chart 
5. 1 line charts

```python
import pandas as pd

dfa = pd.read_excel('obama-approval-ratings.xls', index_col=0)
```

```python
dfa
```

###### 1. Bar Chart

```python
bar_chart = dfa.plot.bar()
bar_chart
```

###### 2. Stacked Bar Chart

```python
stacked_bar_chart = dfa.plot.bar(stacked=True)
stacked_bar_chart
```

###### 3. Pie Chart

```python
import matplotlib.pyplot as plt

dfa_race_relations = dfa.head(1)
dfa_race_relations2 = dfa_race_relations.transpose()

plt.pie( dfa_race_relations2['Race Relations'],labels=dfa_race_relations2.index, shadow=False)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

###### 4. Donut Chart

```python
plt.pie( dfa_race_relations2['Race Relations'],labels=dfa_race_relations2.index, shadow=False)
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

###### 5. Line Chart

```python
dfb = pd.read_excel("world-population.xlsm", index_col= False)

lines = dfb.plot.line(x='Year', y='Population')
lines
```
