###### 2. Spatial Chart

```python
import folium
import matplotlib.pyplot as plt
```

```python
m = folium.Map(location=[0,0], zoom_start=2)
 
# I can add marker one by one on the map
for i in range(0,len(costco)):
   folium.Circle(
      location=[costco.iloc[i]['Latitude'], costco.iloc[i]['Longitude']],
      popup=costco.iloc[i]['City'],
      radius=100, #data.iloc[i]['value']*10000,
      color='crimson',
      fill=True,
      fill_color='crimson'
   ).add_to(m)
 
m
```

###### 3. Contour Chart

```python
costco_2 = costco.groupby(['Latitude', 'Longitude']).agg(['count'])
```

```python
cs = plt.contourf(costco_2)#, levels=[10, 30, 50],
    #colors=['#808080', '#A0A0A0', '#C0C0C0'], extend='both')
cs.cmap.set_over('red')
cs.cmap.set_under('blue')
cs.changed()

```
