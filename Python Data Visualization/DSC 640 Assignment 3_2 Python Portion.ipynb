```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tv_sizes = pd.read_csv('ex3-2/tv_sizes.txt', sep="\t")
birth_rates = pd.read_csv('ex3-2/birth-rate.csv', sep=",")
birth_rates_yrly = pd.read_csv('ex3-2/birth-rates-yearly.csv', sep=",")
crime_rates = pd.read_csv('ex3-2/crimerates-by-state-2005.csv', sep=",")
life_exp = pd.read_csv('ex3-2/life-expectancy.csv', sep=",")
```

```python
tv_sizes.head()
```

###### 1. Scatterplot

```python
plt.plot( 'year', 'rate', data=birth_rates_yrly, linestyle='none', marker='o')
plt.show()
```

###### 2. Bubble Chart

```python
plt.scatter('population', 'aggravated_assault', s='murder',data=crime_rates, alpha=0.5)
plt.show()
```

###### 3. Density Plot

```python
sns.kdeplot(tv_sizes['size'], shade=True)
```
