```python
from selenium import webdriver
from time import sleep
from bs4 import BeautifulSoup
import urllib
from urllib.request import urlopen
import requests

url = "https://www.imdb.com/movies-in-theaters/?ref_=nv_tp_inth_1"

#browser = webdriver.Firefox()
#browser.get(url)

#browser.maximize_window()

page = requests.get(url)

bs = BeautifulSoup(page.content, "lxml")
#ta_divs = bs.find_all("h6")
odd_divs = bs.find_all("div",class_="list_item odd")
even_divs = bs.find_all("div",class_="list_item even")
Director_divs = bs.find_all("dic",class_="txt-block")
```

```python
#1 Title

titles_of_movies = []

for titleo in odd_divs:
    for titleo in titleo.find_all('h4'):
        srcso = titleo.get_text()
        titles_of_movies.append(srcso)

for titlee in even_divs:
    for titlee in titlee.find_all('h4'):
        srcse = titlee.get_text()
        titles_of_movies.append(srcse)
```

```python
#2 Title_Link

title_link = []

for tlinko in odd_divs:
    for to in tlinko.find_all('h4'):
        for to2 in to.find_all('a'):
            s = to2.get('href')
            surl = [url+s]
            title_link.append(surl)

for tlinke in even_divs:
    for te in tlinke.find_all('h4'):
        for te2 in te.find_all('a'):
            s = te2.get('href')
            surl = [url+s]
            title_link.append(surl)
```

```python
#3 Description
about = []

for abouto in odd_divs:
    o = [o2.get_text() for o2 in abouto.find_all("div",class_="outline")]
    about.append(o)

for aboute in even_divs:
    e = [e2.get_text() for e2 in aboute.find_all("div",class_="outline")]
    about.append(e)
```

```python
#4 Poster Link
img_link = []

for linko in odd_divs:
    srcso = [img['src'] for img in linko.find_all('img')[:1]]
    img_link.append(srcso)

for linke in even_divs:
    srcse = [img['src'] for img in linke.find_all('img')[:1]]
    img_link.append(srcse)
    
```

```python
#5 Movie Length

runtime = []

for runo in odd_divs:
    for x in runo.find_all('p',class_='cert-runtime-genre'):
        s = x.find_all("time")
        for s2 in s:
            s3 = [s2.get_text()]
            runtime.append(s3)

for rune in even_divs:
    for x in rune.find_all('p',class_='cert-runtime-genre'):
        s = x.find_all("time")
        for s2 in s:
            s3 = [s2.get_text()]
            runtime.append(s3)
```

```python
#5 Genre(s) of Movie
genre = []

for runo in odd_divs:
    for x in runo.find_all('p',class_='cert-runtime-genre'):
        s = x.find_all("span")
        g=[]
        for s2 in s:
            s3 = s2.get_text()
            g.append(s3)
        genre.append(g)

for rune in even_divs:
    for x in rune.find_all('p',class_='cert-runtime-genre'):
        s = x.find_all("span")
        g=[]
        for s2 in s:
            s3 = s2.get_text()
            g.append(s3)
        genre.append(g)
```

```python
#6 Director
director = []

for do in odd_divs:
    for do2 in do.find_all('div',class_='txt-block'):
        for x in do2.find_all('span'):
            d = [x.get_text().replace('\n','').replace('\r','').strip()]
            director.append(d)

for de in even_divs:
    for de2 in de.find_all('div',class_='txt-block'):
        for x in de2.find_all('span'):
            d = [x.get_text().replace('\n','').replace('\r','').strip()]
            director.append(d)
```

```python
#8 Director_link

director_link = []

for do in odd_divs:
    for do2 in do.find_all('div',class_='txt-block'):
        for do3 in do2.find_all('span'):
            for x in do3.find_all('a')[:1]:
                d = x.get('href')
                durl = [url+d]
                director_link.append(durl)

for de in even_divs:
    for de2 in de.find_all('div',class_='txt-block'):
        for de3 in de2.find_all('span'):
            for x in de3.find_all('a')[:1]:
                d = x.get('href')
                durl = [url+d]
                director_link.append(durl)
```

```python
#9 stars

stars = []

for runo in odd_divs:
    for x in runo.find_all('div',class_='txt-block'):
        s = x.find_all("a")
        g=[]
        for s2 in s:
            s3 = s2.get_text()
            g.append(s3)
        if not g:
            continue
        elif g in director:
            continue
        else:
            stars.append(g)

for rune in even_divs:
    for x in rune.find_all('div',class_='txt-block'):
        s = x.find_all("a")
        g=[]
        for s2 in s:
            s3 = s2.get_text()
            g.append(s3)
        if not g:
            continue
        elif g in director:
            continue
        else:
            stars.append(g)
        
```

```python
#10 stars_link

stars_link = []

for do in odd_divs:
    for do2 in do.find_all('div',class_='txt-block'):
        for span_tag in do2.findAll('span'):
            span_tag.replace_with('')
        g=[]
        for x in do2.find_all('a'):
            d = x.get('href')
            durl = url+d
            g.append(durl)
        if not g:
            continue
        else:
            stars_link.append(g)
            
for de in even_divs:
    for de2 in de.find_all('div',class_='txt-block'):
        for span_tag in de2.findAll('span'):
            span_tag.replace_with('')
        g=[]
        for x in de2.find_all('a'):
            d = x.get('href')
            durl = url+d
            g.append(durl)
        if not g:
            continue
        else:
            stars_link.append(g)
```

```python
#11 Metascore

metascore = []

for abouto in odd_divs:
    for oo in abouto.find_all('h4'):
        title = oo.get_text()

    for o2 in abouto.find_all("div",class_="rating_txt"):
        for o3 in o2.find_all("span",class_="metascore mixed"):
            o = title+';'+o3.get_text().strip()
            metascore.append(o)

for aboute in even_divs:
    for ee in aboute.find_all('h4'):
        title = ee.get_text()

    for e2 in aboute.find_all("div",class_="rating_txt"):
        for e3 in e2.find_all("span",class_="metascore mixed"):
            e = title+';'+e3.get_text().strip()
            metascore.append(e)
```

```python
#12 rating_out_of_10

rating_out_of_10 = []

for abouto in odd_divs:
    for oo in abouto.find_all('h4'):
        title = oo.get_text()

    for o2 in abouto.find_all("span",class_="rating-rating"):
            o = title+';'+o2.get_text().strip()
            rating_out_of_10.append(o)

for aboute in even_divs:
    for ee in aboute.find_all('h4'):
        title = ee.get_text()

    for e2 in aboute.find_all("span",class_="rating-rating"):
            e = title+';'+e2.get_text().strip()
            rating_out_of_10.append(e)
```

```python
#13 Year Movie Made
title_year = []

for titleo in odd_divs:
    for titleo2 in titleo.find_all('h4'):
        srcsoo = titleo2.get_text()
        a = srcsoo.strip().split('(')[1]
        b = [a.strip().split(')')[0]]
        title_year.append(b)

for titlee in even_divs:
    for titlee2 in titlee.find_all('h4'):
        srcsee = titlee2.get_text()
        a = srcsee.strip().split('(')[1]
        b = [a.strip().split(')')[0]]
        title_year.append(b)


```

```python
#14 The Designation of the Audience

audience_des = []

for linko in odd_divs:
    for oo in linko.find_all('h4'):
        title = oo.get_text()
        
    for img in linko.find_all('p',class_='cert-runtime-genre'):
        for img_r in img.find_all('img',class_='absmiddle certimage'):
            srcso = img_r.get("title")
            osrcso = title+';'+srcso
            audience_des.append(osrcso)

for linke in even_divs:
    for ee in linke.find_all('h4'):
        title = ee.get_text()
        
    for img in linke.find_all('p',class_='cert-runtime-genre'):
        for img_r in img.find_all('img',class_='absmiddle certimage'):
            srcse = img_r.get("title")
            esrcse = title+';'+srcse
            audience_des.append(esrcse)

```

```python
#15 Trailer Link
trailer_link = []

for tlinko in odd_divs:
    for to in tlinko.find_all(class_='btn2 btn2_simple small title-trailer video-modal'):
        s = to.get('href')
        surl = [url+s]
        trailer_link.append(surl)

for tlinke in even_divs:
    for te in tlinke.find_all(class_='btn2 btn2_simple small title-trailer video-modal'):
        s = te.get('href')
        surl = [url+s]
        trailer_link.append(surl)

```

###### In this section, we will take the previously scvraped fields and put them into a formatted dataset.

```python
columns = ['Title', 'Title_Link', 'Description', 'Poster_Link', 'Director', 'Director_link', 
           'run_time', 'genre', 'stars', 'stars_links', 
           'movie_year', 'trailer']

columns = [item.strip().lower() for item in columns]

scraped_lists = [titles_of_movies,title_link, about, img_link, director, director_link, 
                 runtime, genre, stars,stars_link,title_year,trailer_link]


columns2 = ['Title','Metascore','rating_out_of_10','audience_des']

columns2 = [item.strip().lower() for item in columns2]

scraped_lists2 = [metascore,metascore,rating_out_of_10,audience_des]
```

```python
imdb_records = []
k = len(scraped_lists[0])

for x in range (0,k):
    lst2 = [item[x] for item in scraped_lists]
    imdb = dict(zip(columns,lst2))
    imdb_records.append(imdb.copy())
```

```python
imdb_meta_recs = []
metascore_dict_list = {}

k = len(metascore)
for x in range(0,k):
    metascore_dict_list["title"] = metascore[x].split(';')[0]
    metascore_dict_list["metascore"] = metascore[x].split(';')[1]
    imdb_meta_recs.append(metascore_dict_list.copy())
    
imdb_10_recs = []
s10_dict_list = {}



k = len(rating_out_of_10)
for x in range(0,k):
    s10_dict_list["title"] = rating_out_of_10[x].split(';')[0]
    s10_dict_list["rating_out_of_10"] = rating_out_of_10[x].split(';')[1]
    imdb_10_recs.append(s10_dict_list.copy())
    
imdb_aud_recs = []
aud_dict_list = {}



k = len(audience_des)
for x in range(0,k):
    aud_dict_list["title"] = audience_des[x].split(';')[0]
    aud_dict_list["audience_des"] = audience_des[x].split(';')[1]
    imdb_aud_recs.append(aud_dict_list.copy())
```

```python
import pandas as pd

imdb_pd = pd.DataFrame(imdb_records)
imdb_meta = pd.DataFrame(imdb_meta_recs)
imdb_10 = pd.DataFrame(imdb_10_recs)
imdb_aud = pd.DataFrame(imdb_aud_recs)
```

```python
result = pd.merge(imdb_pd,imdb_meta[['title','metascore']],on='title', how='left')
result = pd.merge(result,imdb_10[['title','rating_out_of_10']],on='title', how='left')
result = pd.merge(result,imdb_aud[['title','audience_des']],on='title', how='left')
```
